```python
x = np.linspace(-np.pi, np.pi, 20)
cos_x = np.cos(x)
sin_x = np.sin(x)
fig = plt.figure(figsize=(7,5))
ax = fig.add_axes([0, 0, 1, 1])

ax1 = ax.plot(x, cos_x, marker='o', c='c', linewidth=2)
ax2 = ax.plot(x, sin_x, marker='*', c='r', linewidth=2)

ax.legend(labels=['cos(x)', 'sin(x)'])
ax.grid(alpha=0.3)

ax.set_title('Графики тригонометрических функций')

plt.show()
```

```python
from matplotlib.ticker import AutoMinorLocator

x = np.linspace(-np.pi, np.pi, 30)
y = 2 * np.sin(x)

fig = plt.figure(figsize=(6,4))
ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, y, c='royalblue', linewidth=3, marker='o')

ax.xaxis.set_minor_locator(AutoMinorLocator(5))
ax.yaxis.set_minor_locator(AutoMinorLocator(2))

ax.tick_params(which='major', pad=15, axis='x')
ax.tick_params(which='major', pad=10, axis='y')
ax.tick_params(which='minor', labelsize=7, grid_alpha=0.2, labelrotation=90, axis='x')
ax.xaxis.set_minor_formatter('{x:.1f}')
ax.tick_params(which='minor', labelsize=7, grid_alpha=0.2, axis='y')
ax.yaxis.set_minor_formatter('{x:.2f}')
ax.grid(which='both')

plt.show()
```

```python
from matplotlib.gridspec import GridSpec
x = np.linspace(-np.pi, np.pi, 30)
y1 = np.sin(x)
y2 = 0.5 * np.sin(x)
y3 = 2 * np.sin(x)

fig = plt.figure(figsize=(10, 5))

widths = [2, 2, 1]
heights = [1, 2, 2]
gs = GridSpec(nrows=3, ncols=3, figure=fig, width_ratios=widths, height_ratios=heights)
ax1 = fig.add_subplot(gs[0,0])
ax1.plot(x, y1, 'y')
ax2 = fig.add_subplot(gs[0,1])
ax2.plot(x, y2, 'g')
ax3 = fig.add_subplot(gs[1,:])
ax3.plot(x, y3, 'm')

plt.tight_layout()
plt.show()
```

```python
from sklearn.datasets import load_iris

iris_dataset = load_iris()
df = pd.DataFrame(iris_dataset['data'], columns=iris_dataset['feature_names'])
df['target'] = iris_dataset['target']
df['target'] = df['target'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})

display(df)
```

```python
x = df['petal length (cm)']
y = df['petal width (cm)']

plt.scatter(x, y)

plt.xlabel('Длина лепестка')
plt.ylabel('Ширина лепестка')

plt.show()
```

```python
x = 'petal length (cm)'
y = 'petal width (cm)'

groups = df.groupby('target')

for name, group in groups:
    plt.scatter(group[x], group[y], label=name)

plt.legend()
plt.xlabel('Длина лепестка')
plt.ylabel('Ширина лепестка')

plt.show()
```

```python
import numpy as np

np.random.seed(12)
data = []

for _ in range(100):
    data.append(np.random.randint(100))
plt.hist(x=data, bins=20, edgecolor='w')

plt.show()
```

```python
from matplotlib.gridspec import GridSpec

fig = plt.figure(figsize=(10, 5))

widths = [1, 1, 1]
heights = [1, 2, 2]

gs = GridSpec(nrows=3, ncols=3, figure=fig, width_ratios=widths, height_ratios=heights)

ax1 = fig.add_subplot(gs[0:2,0])
ax1.hist(x=data, bins=10, edgecolor='w', color='olive', alpha=0.6)
ax1.grid(axis='y', alpha=0.3)
ax2 = fig.add_subplot(gs[0:2,1])
ax2.hist(x=data, bins=20, edgecolor='w', color='g', alpha=0.6)
ax3 = fig.add_subplot(gs[0:2,2])
ax3.hist(x=data, bins=30, edgecolor='w', orientation='horizontal', color='b', alpha=0.6)
ax4 = fig.add_subplot(gs[2:,:])
ax4.plot(data, color='r', alpha=0.6, linewidth=4)
ax4.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()
```
